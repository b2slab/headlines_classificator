```python
import pickle
import itertools
import os
import glob
```

```python
# Get current path
cpath = os.getcwd()
```

```python
all_headlines = []
for filee in os.listdir(cpath):
    if filee in glob.glob('*.pckl'):
        with open(filee, "rb") as input_file:
            e = pickle.load(input_file)
        all_headlines.append(e)
```

```python
# Grouping all dictionaries into one
dd = {}

for k in all_headlines[0]: 
    dd[k] = [d[k] for d in all_headlines]
```

```python
for key, value in dd.items(): 
    merged = list(itertools.chain.from_iterable(value))
    dd[key] = merged
```
