```python
# Grouping all dictionaries into one
dd = {}

for k in es[0]: 
    dd[k] = [d[k] for d in es]
```

```python
for key, value in dd.iteritems(): 
    merged = list(itertools.chain.from_iterable(value))
    dd[key] = merged
```
